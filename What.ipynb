**29.create function to calculate range of age**

```python
def range_age1(dataframe):
    #loop through dataframe
    for index, row in dataframe.iterrows():
        # if age is greater than 32 then insert 'Almost Retired' into new column
        if row['age'] >= 32:
            dataframe.loc[index, 'age_range'] = 'Almost Retired'
        # if age is geater than or equal to 25 then insert 'Prime' into new column
        elif row['age'] >= 25:
            dataframe.loc[index, 'age_range'] = 'Prime'
        # if age is less than 25 then insert 'Rookies' into new column
        else:
            dataframe.loc[index, 'age_range'] = 'Rookies'
```

หรือ

```python
def range_age2(dataframe):
    # >32 = Almost Retired
    dataframe.loc[dataframe['age'] >= 32, 'age_range'] = 'Almost Retired'
    # >=25 and <32 = Prime
    dataframe.loc[(dataframe['age'] >= 25) & (dataframe['age'] < 32), 'age_range'] = 'Prime'
    # <25 = Rookies
    dataframe.loc[dataframe['age'] < 25, 'age_range'] = 'Rookies'
```
